```python
import pandas as pd
```

```python
#Results from matrix multiplications of 100X100, 125X125, 150X150, 175X175 and 200X200 one one working node
withOneNode = {'100':41, '125':47, '150':58,'175':72,'200':90}
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
plt.title('One Working Node with Multiple Multiplications', fontsize=20)
plt.xlabel('Size of Matrices', fontsize=15)
plt.ylabel('Execution Time in Seconds', fontsize=15)
x, = plt.plot(list(withOneNode.keys()), list(withOneNode.values()))
plt.plot(list(withOneNode.keys()), list(withOneNode.values()), 'ro')
plt.legend([x],['Preliminary'], loc=1)
```

```python
#Results from matrix multiplication of 500X500 one one, two, three and four working nodes
mat500 = {'1': 34, '2': 19, '3':11, '4': 11}
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
plt.title('Size 500 Multiplication with Multiple Nodes', fontsize=20)
plt.xlabel('No:of Nodes', fontsize=15)
plt.ylabel('Execution Time in Minutes', fontsize=15)
x, = plt.plot(list(mat500.keys()), list(mat500.values()))
plt.plot(list(mat500.keys()), list(mat500.values()), 'ro')
plt.legend([x],['Preliminary'], loc=1)
```

```python
withOneNodeOld = {'100':41, '125':47, '150':58,'175':72,'200':90}
withOneNodeNew = {'100':28, '125':27, '150':26,'175':28,'200':26}
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
plt.title('One Working Node with Multiple Multiplications', fontsize=20)
plt.xlabel('Size of Matrices', fontsize=15)
plt.ylabel('Execution Time in Seconds', fontsize=15)
x, = plt.plot(list(withOneNodeOld.keys()), list(withOneNodeOld.values()))
y, = plt.plot(list(withOneNodeNew.keys()), list(withOneNodeNew.values()))
plt.plot(list(withOneNodeOld.keys()), list(withOneNodeOld.values()), 'ro')
plt.plot(list(withOneNodeNew.keys()), list(withOneNodeNew.values()), 'ro')
plt.legend([x,y],['Preliminary','Final'], loc=1)
```
